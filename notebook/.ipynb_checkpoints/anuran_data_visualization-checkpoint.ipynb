```python
import numpy as np
import os

DATA_PATH = 'Anuran_Calls_(MFCCs)'
```

```python
import pandas as pd

def load_csv(path = DATA_PATH):
    csv_path = os.path.join(path, 'Frogs_MFCCs.csv')
    return pd.read_csv(csv_path)
```

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# example for values of families
family_data = Y.loc[:,'Family']

# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(family_data)
print(integer_encoded)
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```
